```python
df=pd.read_csv('FinalADME..csv')
df
```

Data preparation


Data separation as X and y

```python
y= y = df['MLOGP']
y
X = df.drop('MLOGP', axis=1)
X
```

Data splitting

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

Model Building

Linear Regression

Training the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

Random Forest

Training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

SVM  Regression model

```python
!pip install scikit-learn
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
svr = SVR() # Support Vector Machine (SVR)
# Initialize the SVR model
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)

# Train the model
svr_model.fit(X_train, y_train)
```

```python
# Make predictions
y_train_pred_svr = svr_model.predict(X_train)
y_test_pred_svr = svr_model.predict(X_test)

# Calculate performance metrics
mse_train_svr = mean_squared_error(y_train, y_train_pred_svr)
mse_test_svr = mean_squared_error(y_test, y_test_pred_svr)
rmse_train_svr = np.sqrt(mse_train_svr)
rmse_test_svr = np.sqrt(mse_test_svr)
mae_train_svr = mean_absolute_error(y_train, y_train_pred_svr)
mae_test_svr = mean_absolute_error(y_test, y_test_pred_svr)
r2_train_svr = r2_score(y_train, y_train_pred_svr)
r2_test_svr = r2_score(y_test, y_test_pred_svr)
```

```python
# Print the results
print("SVM Model (SVR) Performance:")
print("Training Metrics:")
print(f"  MSE: {mse_train_svr}")
print(f"  RMSE: {rmse_train_svr}")
print(f"  MAE: {mae_train_svr}")
print(f"  R^2: {r2_train_svr}")

print("\nTesting Metrics:")
print(f"  MSE: {mse_test_svr}")
print(f"  RMSE: {rmse_test_svr}")
print(f"  MAE: {mae_test_svr}")
print(f"  R^2: {r2_test_svr}")
```

```python
svr_results = pd.DataFrame(['Support Vector Machine', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
svr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
svr_results
```

Neural Network (NN) model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Define the Neural Network model
model = Sequential()

# Add layers to the model
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))  # Input layer + first hidden layer
model.add(Dense(32, activation='relu'))  # Second hidden layer
model.add(Dense(16, activation='relu'))  # Third hidden layer
model.add(Dense(1))  # Output layer (single neuron for regression)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)

# Make predictions
y_train_pred_nn = model.predict(X_train).flatten()
y_test_pred_nn = model.predict(X_test).flatten()
```

```python
# Make predictions
y_train_pred_nn = model.predict(X_train).flatten()
y_test_pred_nn = model.predict(X_test).flatten()

# Calculate performance metrics
mse_train_nn = mean_squared_error(y_train, y_train_pred_nn)
mse_test_nn = mean_squared_error(y_test, y_test_pred_nn)
rmse_train_nn = np.sqrt(mse_train_nn)
rmse_test_nn = np.sqrt(mse_test_nn)
mae_train_nn = mean_absolute_error(y_train, y_train_pred_nn)
mae_test_nn = mean_absolute_error(y_test, y_test_pred_nn)
r2_train_nn = r2_score(y_train, y_train_pred_nn)
r2_test_nn = r2_score(y_test, y_test_pred_nn)
```

```python
# Print the results
print("Neural Network Model Performance:")
print("Training Metrics:")
print(f"  MSE: {mse_train_nn}")
print(f"  RMSE: {rmse_train_nn}")
print(f"  MAE: {mae_train_nn}")
print(f"  R^2: {r2_train_nn}")

print("\nTesting Metrics:")
print(f"  MSE: {mse_test_nn}")
print(f"  RMSE: {rmse_test_nn}")
print(f"  MAE: {mae_test_nn}")
print(f"  R^2: {r2_test_nn}")
```

```python
NN_results = pd.DataFrame(['Nrural Network model', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
NN_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
NN_results
```
